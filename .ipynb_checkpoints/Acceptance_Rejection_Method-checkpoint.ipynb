```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Density of |Z| with Z following the normal distribution

def f(x):
    return (2/np.sqrt(2*np.pi))*np.exp(-(x**2)/2)


# Expo density with rate 1
def g(x):             
    return np.exp(-x)
 
```

```python
xx=np.linspace(-3,3,100)


plt.plot(xx, f(xx),color= 'red')
plt.plot(xx, g(xx), color='blue')
plt.plot(-xx, g(xx), color='blue')
plt.ylim(0,1)
```

```python
# h(x) = f(x)/g(x) = (np.sqrt(2/np.pi))*exp{(x - x**2)/2} 

def h(x):
    return f(x)/g(x)

# h'(x) = 0 implies c = max{f(x)/g(x)} = np.sqrt(2e/np.pi)
```

```python
# c=f(1)/g(1)
c=np.sqrt(2*np.exp(1)/np.pi)
c
```

```python
plt.plot(xx, h(xx), color ='blue')
plt.plot(1,c ,"o",markersize=10)
# plt.ylim(0,1)
```

```python

plt.plot(xx, f(xx),color= 'red')
plt.plot(xx, c*g(xx), color='blue')
plt.plot(-xx, c*g(xx), color='blue')
plt.ylim(0,1.3)
```

```python
np.random.uniform(low = 0.0, high = 1.0, size = None)
```

```python
N=100000
sample = []
```

```python
for i in range(N):
    while True:
        y = -np.log(np.random.uniform(low = 0.0, high = 1.0, size = None)) # Inverse CDF
        U = np.random.uniform(low = 0.0, high = 1.0, size = None)
        if U <= (1/c)*h(y):
            if np.random.uniform(low = 0.0, high = 1.0, size = None) <= 0.5 :
                sample.append(y)
            else :
                sample.append(-y)
            break 

    
```

```python
len(sample)
```

```python
sample=np.array(sample)
```

```python
plt.plot(sample,"o")
```

```python
plt.hist(sample, bins=100,density= True)
plt.plot(xx, 0.5*f(xx),color= 'black')
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(xx, f(xx),color= 'red', label = 'f(x)')

plt.plot(xx, 0.5*f(xx),color= 'black',label = 'Standard Normal distribution')
plt.plot(xx, c*g(xx), color='blue',label = 'c*g(x)')
plt.plot(-xx, c*g(xx), color='blue')
plt.hist(sample, bins=100,density= True)
plt.ylim(0,1.3)
plt.legend()
plt.show()
```
