Computation of
$$S_k^{ij}=(C_{d_0d_0})^{ij} + \int_a^b  \left( \int_a^b  G_k^i(w').C_{p_0p_0}(w',w^{''}).G_k^j(w{''})  dw^{''}\right) dw'=(C_{d_0d_0})^{ij} + \int_a^b  G_k^i(w') \left( \int_a^b  C_{p_0p_0}(w',w^{''}).G_k^j(w{''})  dw^{''}\right) dw'$$
