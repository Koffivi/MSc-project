```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Density of |Z| with Z following the normal distribution

def f(x):
    return (2/np.sqrt(2*np.pi))*np.exp(-(x**2)/2)


# Expo density with rate 1
def g(x):             
    return np.exp(-x)

def h(x):
    return f(x)/g(x)

c=f(1)/g(1)
c/2
```

```python
def sampling(sig,mu,N):
    sample = []
    for i in range(N):
        while True:
            U1=np.random.uniform(low = 0.0, high = 1.0, size = None)
            y = -np.log(U1) # Inverse CDF
            U2 = np.random.uniform(low = 0.0, high = 1.0, size = None)
            if U2 <= (1/c)*h(y):
                if np.random.uniform(low = 0.0, high = 1.0, size = None) <= 0.5 :
                    sample.append(sig*y + mu)
                else :
                    sample.append(sig*(-y)+mu)
                break 
    return sample
    
```

```python
Sigma = 0.1
Mu = 4
N=10000
```

```python
XX = np.random.normal(Mu,Sigma,N)

generated_sample=sampling(Sigma,Mu,N)
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
plt.hist(generated_sample, bins=100,density= True,color='red')
plt.title("Generated sample with mu={},sigma={} and size={}".format(Mu,Sigma,N))
# plt.ylim(0,1)


plt.subplot(1,2,2)
plt.hist(XX, bins=100,density= True)
plt.title("np.random.normal({},{},{})".format(Mu,Sigma,N))
# plt.ylim(0,1)


# plt.legend()
plt.show()
```
