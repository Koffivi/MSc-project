```python
def f(x):
    return (2/np.sqrt(2*np.pi))*np.exp(-(x**2)/2)


def g(x):             
    return np.exp(-x)

def h(x):
    return f(x)/g(x)

c=np.sqrt(2*np.exp(1)/np.pi)

 
```

```python
def F(N, mu, sigma):
    
    
    sample = []
    for i in range(N):
        while True:
            u1=np.random.uniform(low = 0.0, high = 1.0, size = None)
            y = -np.log(u1) # Inverse CDF
            U = np.random.uniform(low = 0.0, high = 1.0, size = None)
            if U <= (1/c)*h(y):
                if np.random.uniform(low = 0.0, high = 1.0, size = None) <= 0.5 :
                    sample.append(mu+y*sigma)
                else :
                    sample.append(mu-y*sigma)
                break
                    
    return sample
```

$f(x)=\dfrac{1}{\sigma\sqrt{2\pi}} \exp\left[-\dfrac{1}{2}\dfrac{(x-\mu)^2}{\sigma^2}\right]$



a. Sample of $x$ which follow a normal distribution with mean 4 and variance 0.001

$ x\sim \mathcal{N}(4, 0.001) \implies x=4+(0.001)^{\frac{1}{2}} Z $ ,$~~~~ $  where $Z \sim \mathcal{N}(0,1)$

```python
N=10000
mu=4
sigma=np.sqrt(.001)
FX=F(N,mu,sigma)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1) 
plt.hist(FX,bins=100,density=True)
plt.title('Generated Sample mean=4 and variance=0.001')

plt.subplot(1,2,2) 
plt.hist(np.random.normal(4,np.sqrt(.001),10000),bins=100,density=True)
plt.title('Python built-in function mean=4 and variance=0.001')
plt.show()
```

b. Sample of $x$ which follow a normal distribution with mean -10 and variance 0.001

```python
mu2=4
sig2=0.01
```

```python
def f2(x,mu,sig):
    return (1/np.sqrt(2*np.pi*(sig**2)))*np.exp(-((x-mu)**2)/(2*sig**2))

def g2(x):
    return np.exp(-x)

def h2(x,mu,sig):
    return f2(x,mu,sig)/g2(x)
```

```python
c2 = f2(mu2 + sig2**2,mu2,sig2)/g2(mu2+sig2**2)
print(c2)

h2(mu2 + sig2**2,mu2,sig2)
```

```python
xx= np.linspace(-4,10,10000)
plt.plot(xx,f2(xx,mu2,sig2),color = 'red')
plt.plot(xx,c2*g2(xx))
plt.plot(-xx,c2*g2(xx))
plt.ylim(0,100)
```

```python
N=1000
sample = []

for i in range(N):
    while True:
        U1=np.random.uniform(low = 0.0, high = 1.0, size = None)
        y = -np.log(U1) # Inverse CDF
        U2 = np.random.uniform(low = 0.0, high = 1.0, size = None)
        if U2 <= (c2**(-1))*(f2(y,mu2,sig2)/g2(y)):
            sample.append(y)
            break 
```
